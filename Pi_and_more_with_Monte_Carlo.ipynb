## What is the Monte Carlo Method?

Imagine you are **blindfolded** and throwing darts at a wall. On the wall there is a dartboard — a circular target — but the circle is hung inside a **square frame**. You can't see anything, so your darts land at completely random spots on the square frame.

After throwing many, many darts, you ask your friend to count two things:
- How many darts landed **inside the circle**?
- How many darts landed **anywhere on the square** (inside or outside the circle)?

Here is the key insight: the **ratio** of darts inside the circle to total darts tells you something about the **area of the circle relative to the area of the square**. No tape measure needed — just counting random darts!

This is the essence of the **Monte Carlo method**: use randomness and counting to estimate things that are hard to calculate directly.

### Where did the name come from?

The method is named after the **Monte Carlo Casino** in Monaco, a tiny country on the French Riviera famous for gambling. Gambling is all about randomness and probability — exactly what this method uses. The name was coined by mathematician Stanislaw Ulam in the 1940s while working on the Manhattan Project.

### When is Monte Carlo used?

Monte Carlo is used when problems are **too complex to solve with a direct formula**. Instead of finding the exact answer mathematically, we:

1. **Generate** random trials (throw darts, flip coins, simulate experiments)
2. **Count** how many trials satisfy the condition we care about
3. **Use the ratio** to estimate the answer

The golden rule: **more trials = more accurate answer**. A thousand darts give a better estimate than ten darts. A million give a better estimate than a thousand. The more we simulate, the closer we get to the truth.

## Part 1 — Estimating π with a Dartboard

Here is a beautiful application of Monte Carlo: we can estimate the value of **π (pi)** — one of the most famous numbers in mathematics — just by throwing random darts!

### The Setup

Imagine a **square** with sides of length 1 (so its area = 1 × 1 = 1).  
Inside that square, draw a **quarter circle** with radius 1, centred at the bottom-left corner.

- Area of the full square = 1
- Area of the quarter circle = π × r² / 4 = π × 1² / 4 = **π/4**

### The Dart Throwing

Now throw darts randomly at the square. Each dart lands at a random (x, y) point where both x and y are between 0 and 1.  
A dart is **inside the quarter circle** if the distance from the origin is ≤ 1, i.e. if **x² + y² ≤ 1**.

### The Formula

Since the darts are random and uniformly spread:

$$\frac{\text{darts inside circle}}{\text{total darts}} \approx \frac{\text{area of quarter circle}}{\text{area of square}} = \frac{\pi/4}{1} = \frac{\pi}{4}$$

So:

$$\pi \approx 4 \times \frac{\text{points inside circle}}{\text{total points}}$$

### Why Does This Work?

If darts truly land randomly and uniformly, the **fraction** that land inside the circle must equal the **fraction of the area** the circle occupies. We know the square's area is 1 and the quarter circle's area is π/4, so the ratio must be π/4. Multiply by 4 and we get π. No calculus, no memorised formula — just counting darts!

```python
# Suppress warnings to keep the output clean for reading
import warnings
warnings.filterwarnings('ignore')

import random
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import seaborn as sns

# Use a clean, white grid style for all our plots
sns.set_style('whitegrid')

# random_state=42 ensures we get the same results every time we run this notebook
# This is important so your classmates see the exact same output as you
random.seed(42)
np.random.seed(42)

print("All libraries loaded successfully!")
print("Random seed set to 42 — results will be reproducible.")
```

```python
# Pi estimation accuracy table
# We test with increasing numbers of random points to see how accuracy improves

# Reset the seed before this experiment so results are consistent
np.random.seed(42)

# The number of random points to test at each level
point_counts = [100, 1_000, 10_000, 100_000, 1_000_000]

print("Estimating \u03c0 with Monte Carlo — Accuracy vs Number of Points")
print("-" * 70)
print(f"{'Total Points':>15} | {'Estimated \u03c0':>12} | {'Real \u03c0':>10} | {'Error':>10} | {'% Error':>8}")
print("-" * 70)

for num_points in point_counts:
    # Generate num_points random (x, y) pairs, each between 0 and 1
    # numpy lets us generate many at once — much faster than a Python loop
    x_vals = np.random.uniform(0, 1, num_points)
    y_vals = np.random.uniform(0, 1, num_points)

    # Calculate the distance from the origin for every point
    distances = np.sqrt(x_vals**2 + y_vals**2)

    # Count how many points are inside the quarter circle (distance <= 1)
    num_inside = np.sum(distances <= 1)

    # Apply the Monte Carlo pi formula: pi ≈ 4 × (inside / total)
    estimated_pi = 4 * num_inside / num_points

    # Calculate absolute and percentage error compared to the real pi
    abs_error = abs(math.pi - estimated_pi)
    pct_error = (abs_error / math.pi) * 100

    print(f"{num_points:>15,} | {estimated_pi:>12.6f} | {math.pi:>10.6f} | "
          f"{abs_error:>10.6f} | {pct_error:>7.4f}%")

print("-" * 70)
print(f"\nThe true value of \u03c0 is {math.pi:.10f}...")
print("Watch how the error shrinks as we use more and more random points!")
```

### **What do we see here?**

- **With only 100 points**, our estimate of π is quite rough — we might be off by 0.05 or more. Small samples are unreliable.
- **Each time we multiply the number of points by 10**, the error roughly gets about 3× smaller. This is not a coincidence — it follows a mathematical pattern called the "square root rule" of convergence.
- **With 1,000,000 points**, we are usually accurate to 3 or 4 decimal places — impressive for just counting random dots!
- **This pattern** — more trials → better estimate — is the heart of every Monte Carlo method. It is why simulations run for millions of iterations.
- We never solved any equation or looked up any formula for π. **We just counted random dots!**

```python
# Visualise the Monte Carlo dartboard
# We use 3000 points — enough to see the pattern clearly without being too slow
np.random.seed(42)

num_visualization_points = 3000

# Generate random x and y coordinates, each between 0 and 1
x_coords = np.random.uniform(0, 1, num_visualization_points)
y_coords = np.random.uniform(0, 1, num_visualization_points)

# Check which points are inside the quarter circle (distance from origin <= 1)
distances = np.sqrt(x_coords**2 + y_coords**2)
inside_circle = distances <= 1

# Estimate pi using our formula
estimated_pi_visual = 4 * np.sum(inside_circle) / num_visualization_points

fig, ax = plt.subplots(figsize=(8, 8))

# Plot points outside the circle (coral/red)
ax.scatter(x_coords[~inside_circle], y_coords[~inside_circle],
           color='#D64045', alpha=0.4, s=5, label='Outside circle')

# Plot points inside the circle (teal/green)
ax.scatter(x_coords[inside_circle], y_coords[inside_circle],
           color='#2A9D8F', alpha=0.4, s=5, label='Inside circle')

# Draw the quarter circle arc — this is the boundary we are testing against
arc = patches.Arc((0, 0), 2, 2, angle=0, theta1=0, theta2=90,
                  color='#1E3A5F', linewidth=3)
ax.add_patch(arc)

# Draw the square border so the frame is clearly visible
square = patches.Rectangle((0, 0), 1, 1, linewidth=2,
                            edgecolor='#1E3A5F', facecolor='none')
ax.add_patch(square)

ax.set_xlim(-0.05, 1.05)
ax.set_ylim(-0.05, 1.05)
ax.set_aspect('equal')
ax.set_title(f'Monte Carlo Dartboard — Estimating \u03c0\n'
             f'Estimated \u03c0 = {estimated_pi_visual:.5f}  |  Real \u03c0 = {math.pi:.5f}',
             fontsize=14, fontweight='bold')
ax.set_xlabel('X coordinate', fontsize=12)
ax.set_ylabel('Y coordinate', fontsize=12)
ax.legend(fontsize=11, markerscale=3)

plt.tight_layout()
plt.show()

print(f"\nWith {num_visualization_points:,} random darts:")
print(f"  Estimated \u03c0 = {estimated_pi_visual:.5f}")
print(f"  Real \u03c0      = {math.pi:.5f}")
print(f"  Error       = {abs(math.pi - estimated_pi_visual):.5f}")
```

### **Key Insight — The Power of Randomness**

Look at the scatter plot above. You can clearly see that **more teal dots land inside the arc** than coral dots — this is exactly what we expect because the quarter circle takes up about 78.5% of the square (π/4 ≈ 0.785).

- The **ratio of teal dots to all dots** equals π/4 — and multiplying by 4 gives us π.
- **No calculus was needed.** We didn't differentiate or integrate a single function. We just asked: is this dot inside the circle? Yes or no?
- This visualisation shows why Monte Carlo is so powerful — it turns a hard geometry problem into a simple counting problem.

## Part 2 — Coin Flip & The Law of Large Numbers

Have you ever flipped a coin 10 times and gotten 7 heads? That is 70% — way more than the 50% you expected. Does that mean the coin is rigged?

No! It just means **10 flips is not enough** to see the true pattern.

### The Law of Large Numbers

The **Law of Large Numbers** is one of the most important ideas in all of mathematics and statistics. It says:

> *As you repeat an experiment more and more times, the average result gets closer and closer to the true expected value.*

With a fair coin:
- Flip it **10 times** → you might get 70% heads (feels unfair, but it's just random noise)
- Flip it **1,000 times** → you'll probably be between 48% and 52% heads
- Flip it **100,000 times** → you'll be extremely close to exactly 50%

### Why does Monte Carlo rely on this law?

Monte Carlo is **powered by the Law of Large Numbers**. Every Monte Carlo estimate gets better with more trials for exactly this reason — the randomness averages out and the true answer emerges. This is not a coincidence or a hope — it is a mathematical guarantee.

```python
# Coin flip simulation — watching the Law of Large Numbers in action
# Set seed for reproducibility
np.random.seed(42)

# Simulate 10,000 coin flips
# 1 = Heads, 0 = Tails
total_flips = 10_000
coin_flips = np.random.randint(0, 2, total_flips)  # Generates 0s and 1s randomly

# Calculate the running (cumulative) proportion of heads after each flip
# np.cumsum adds up all the heads so far after each flip
cumulative_heads = np.cumsum(coin_flips)
flip_numbers = np.arange(1, total_flips + 1)
running_proportion = cumulative_heads / flip_numbers

# --- Print milestones ---
print("Law of Large Numbers — Coin Flip Simulation")
print("-" * 55)
milestones = [10, 50, 100, 500, 1000, 5000, 10000]
for milestone in milestones:
    proportion = running_proportion[milestone - 1]
    close = abs(proportion - 0.5) < 0.02
    status = '\u2713 close!' if close else 'still converging...'
    print(f"After {milestone:>6,} flips: {proportion:.4f} heads  ({status})")
print("-" * 55)

# --- Plot ---
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(flip_numbers, running_proportion, color='#2A9D8F', linewidth=1.5,
        label='Running proportion of Heads', alpha=0.85)

# Dashed line at the true probability of 0.5
ax.axhline(y=0.5, color='#D64045', linestyle='--', linewidth=2,
           label='True probability = 0.50')

# Shade the region around 0.5 to show the "acceptable zone"
ax.fill_between(flip_numbers, 0.48, 0.52, color='#D64045', alpha=0.08,
                label='\u00b12% band around 0.50')

# Log scale makes it easier to see early flips AND late flips on the same graph
ax.set_xscale('log')
ax.set_xlabel('Number of Coin Flips (log scale)', fontsize=12)
ax.set_ylabel('Proportion of Heads', fontsize=12)
ax.set_title('Law of Large Numbers — Coin Flip Simulation\n'
             'As we flip more coins, the proportion gets closer to 0.50',
             fontsize=14, fontweight='bold')
ax.set_ylim(0.3, 0.7)
ax.legend(fontsize=11)

plt.tight_layout()
plt.show()
```

### **What do we see here?**

- **After just 10 flips**, the line jumps wildly — anywhere from 20% to 80% heads. Small samples are unreliable and misleading.
- **After about 1,000 flips**, the line settles into the shaded ±2% band and stays there. The estimate has stabilised.
- **After 10,000 flips**, the running proportion is extremely close to 0.50 — often within 0.5%.
- This is exactly **why Monte Carlo needs many iterations** to work well. With only 100 dart throws, our π estimate is rough. With 1,000,000 throws, it is excellent.
- **Real-world connection:** Polling companies survey thousands of people (not just 10) before an election for exactly the same reason — small samples are noisy, large samples give reliable estimates.

## Part 3 — The Birthday Paradox

Here is a question that stumps almost everyone the first time they hear it:

> **How many people do you need in a room before there is a better-than-50% chance that two of them share the same birthday?**

Most people guess around 183 — roughly half of 365, thinking "you need about half as many people as there are days in the year."

**The real answer is just 23 people. With only 23 people!**

This surprises everyone — even professional mathematicians find it counterintuitive the first time.

### Why is the guess wrong?

Here is the key insight: you are not asking "does anyone in the room share **MY** birthday?" You are asking "does **any pair** of people in the room share **a** birthday?"

Think about it this way: with 23 people, there are **253 different pairs** of people (23 × 22 / 2 = 253). Each pair has a small but real chance of sharing a birthday. Across 253 pairs, the total probability adds up surprisingly fast!

Instead of working through the complicated probability formula, let's just **simulate it** with Monte Carlo. We will generate random groups of people, give everyone a random birthday, and count how often at least two birthdays match.

```python
# Birthday Paradox — Monte Carlo simulation
np.random.seed(42)

# We'll test group sizes from 2 people up to 60 people
group_sizes = list(range(2, 61))
num_simulations = 10_000  # Run 10,000 trials for each group size

probabilities = []  # Store the estimated probability for each group size

print("Birthday Paradox — Monte Carlo Simulation")
print("-" * 50)
print(f"Running {num_simulations:,} simulations for each group size...")
print()

for group_size in group_sizes:
    shared_birthday_count = 0

    for _ in range(num_simulations):
        # Generate a random birthday (day of year, 1 to 365) for each person
        birthdays = np.random.randint(1, 366, size=group_size)

        # If the number of unique birthdays is less than the group size,
        # at least two people must share a birthday
        if len(set(birthdays)) < group_size:
            shared_birthday_count += 1

    # The probability is the fraction of simulations where a match occurred
    probability = shared_birthday_count / num_simulations
    probabilities.append(probability)

    # Print results for key group sizes of interest
    if group_size in [10, 20, 23, 30, 40, 50, 57]:
        print(f"Group of {group_size:>2} people: {probability:.1%} chance of a shared birthday")

# Find the first group size where probability crosses 50%
crossover_size = next(size for size, prob in zip(group_sizes, probabilities) if prob >= 0.5)
print()
print(f">>> The probability crosses 50% at a group size of: {crossover_size} people!")
print("-" * 50)

# --- Plot ---
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(group_sizes, probabilities, color='#1E3A5F', linewidth=2.5,
        marker='o', markersize=4, label='Probability of shared birthday')

# Mark the 50% threshold line
ax.axhline(y=0.5, color='#D64045', linestyle='--', linewidth=2, label='50% probability')

# Mark the exact crossover point on the chart
crossover_prob = probabilities[crossover_size - 2]  # index offset: group_sizes starts at 2
ax.axvline(x=crossover_size, color='#E76F51', linestyle=':', linewidth=2,
           label=f'Crossover at {crossover_size} people')
ax.scatter([crossover_size], [crossover_prob], color='#E76F51', s=150, zorder=5)
ax.annotate(f'  {crossover_size} people\n  {crossover_prob:.1%} chance',
            xy=(crossover_size, crossover_prob), fontsize=11,
            color='#E76F51', fontweight='bold')

ax.set_xlabel('Number of People in the Room', fontsize=12)
ax.set_ylabel('Probability of a Shared Birthday', fontsize=12)
ax.set_title('The Birthday Paradox — Simulated with Monte Carlo\n'
             "How many people until there's a >50% chance of a shared birthday?",
             fontsize=14, fontweight='bold')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
ax.legend(fontsize=11)

plt.tight_layout()
plt.show()
```

### **What do we see here? — The Paradox Revealed!**

- **Most people guess ~183**, thinking you need about half as many people as there are days in the year. But the simulation proves that only **23 people** gives you a better-than-50% chance!
- **With 57 people**, the simulation shows approximately a **99% chance** that two people share a birthday — almost certain!
- This feels wrong because our intuition focuses on **our own** birthday vs everyone else's. But the question is about **any pair** matching — and 23 people create 253 different pairs, each with a small chance of matching.
- In a **typical class of 30 students**, there is about a **70% chance** that two of your classmates share a birthday. Try it!
- **Real-world cryptography connection:** The Birthday Paradox is the basis for the "birthday attack" in hashing — a technique hackers use to find two different inputs that produce the same hash output. Security engineers study this problem seriously.

## Part 4 — Estimating Area Under a Curve (The Skateboard Ramp)

Here is a new challenge: imagine a **skateboard ramp** built in a skate park. When you look at the ramp from the side, its curved surface follows the mathematical shape **y = x²** — a parabola.

An engineer needs to know the total **area under the ramp** (from x = 0 to x = 1) to figure out how much concrete to pour. In a maths class, this would normally require **calculus (integration)** — a topic most students don't see until Year 11 or 12.

**But we can estimate this area right now using Monte Carlo!**

### The Method

1. Draw a **1 × 1 square** around the ramp (x from 0 to 1, y from 0 to 1)
2. Throw **random points** at the square
3. Count how many land **below the curve** (i.e. where y ≤ x²)
4. Use the ratio:

$$\text{Estimated Area} = \frac{\text{points below curve}}{\text{total points}} \times \text{area of rectangle}$$

Since the rectangle is 1 × 1 = 1, the formula simplifies to:

$$\text{Estimated Area} \approx \frac{\text{points below curve}}{\text{total points}}$$

The **exact answer** (which calculus gives us) is **1/3 ≈ 0.33333...**  
Let's see how close Monte Carlo gets!

```python
# Monte Carlo integration — estimating the area under y = x² from 0 to 1
np.random.seed(42)

# The curve we are studying: y = x squared
# The exact area under this curve from x=0 to x=1 is exactly 1/3 (proven with calculus)
exact_area = 1 / 3

print("Estimating Area Under y = x\u00b2 Using Monte Carlo")
print("-" * 60)
print(f"Exact answer (from calculus): {exact_area:.8f}")
print()

# Test with different numbers of random points
point_counts = [100, 1_000, 10_000, 100_000, 1_000_000]
estimated_areas = []

for num_points in point_counts:
    # Generate random x values and y values, both between 0 and 1
    x_random = np.random.uniform(0, 1, num_points)
    y_random = np.random.uniform(0, 1, num_points)

    # A point is "under the curve" if its y value is less than or equal to x squared
    below_curve = y_random <= x_random**2

    # Estimate area: fraction of points below curve
    # (rectangle area = 1x1 = 1, so no additional scaling is needed)
    area_estimate = np.sum(below_curve) / num_points
    estimated_areas.append(area_estimate)

    error = abs(exact_area - area_estimate)
    print(f"Points: {num_points:>9,} | Estimated Area: {area_estimate:.6f} | "
          f"Exact: {exact_area:.6f} | Error: {error:.6f}")

print("-" * 60)

# --- Visualisation ---
# Use a fresh set of random points just for the plot
np.random.seed(42)
num_viz_points = 5_000
x_viz = np.random.uniform(0, 1, num_viz_points)
y_viz = np.random.uniform(0, 1, num_viz_points)
below_viz = y_viz <= x_viz**2

fig, ax = plt.subplots(figsize=(9, 7))

# Plot points above the curve (coral — do not count toward area)
ax.scatter(x_viz[~below_viz], y_viz[~below_viz], color='#D64045', alpha=0.3,
           s=4, label='Above curve y = x\u00b2')

# Plot points below the curve (teal — these count toward the estimated area)
ax.scatter(x_viz[below_viz], y_viz[below_viz], color='#2A9D8F', alpha=0.3,
           s=4, label='Below curve y = x\u00b2 (counted!)')

# Draw the actual curve y = x squared so we can see it clearly
x_curve = np.linspace(0, 1, 300)
y_curve = x_curve**2
ax.plot(x_curve, y_curve, color='#1E3A5F', linewidth=3, label='Curve: y = x\u00b2')

# Shade the area under the curve to highlight what we are estimating
ax.fill_between(x_curve, y_curve, alpha=0.2, color='#1E3A5F')

ax.set_xlabel('x', fontsize=13)
ax.set_ylabel('y', fontsize=13)
area_estimate_viz = np.sum(below_viz) / num_viz_points
ax.set_title(f'Monte Carlo Integration — Area Under y = x\u00b2\n'
             f'Estimated area = {area_estimate_viz:.4f}  |  Exact area = {exact_area:.4f}',
             fontsize=14, fontweight='bold')
ax.legend(fontsize=11, markerscale=3)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

### **What do we see here?**

- The **teal dots below the curve** form the characteristic "skateboard ramp" parabola shape — you can see the curve's outline emerging from the random scatter!
- The **ratio of teal dots to all dots** gives us approximately **1/3 ≈ 0.3333** — matching the exact calculus answer without ever integrating.
- As we increase the number of points from 100 to 1,000,000, the estimate converges steadily toward the true answer.
- This technique is called **Monte Carlo Integration** and it scales to problems that are genuinely impossible to solve with formulas. In **physics simulations**, engineers estimate volumes of irregular 3D shapes with millions of random points. In **computer graphics**, Monte Carlo integration calculates how light bounces around a scene to produce realistic shadows and reflections — it's the technology inside every modern animated film.

## Part 5 — The Random Walk (The Lost Ant)

Imagine an ant gets lost in the middle of a giant tiled floor. Every second, the ant randomly picks one of four directions — **up, down, left, or right** — and takes one step that way.

The ant has absolutely **no memory** of where it has been before. Each step is completely independent and random. This is called a **Random Walk**.

### The Questions

- Where will the ant end up after 1,000 steps?
- Could it walk in one direction and end up 1,000 steps away from the start?
- On average, how far from the starting tile does the ant end up?

We will simulate many ants at once and discover a beautiful hidden pattern.

### Real-World Random Walks

Random walks are not just about ants! The same mathematics describes:
- **Gas molecules** bouncing around in a room (Brownian motion)
- **Stock prices** fluctuating on the stock market
- **How diseases spread** through a population
- **How AI explores** game boards in reinforcement learning

One of the most beautiful results in all of mathematics is that after **n steps**, the ant ends up, on average, approximately **√n steps** from the start. Let's verify this ourselves!

```python
# 2D Random Walk simulation — plotting lost ants!
np.random.seed(42)

num_ants = 10        # How many ants to simulate and plot individually
num_steps = 1_000    # How many steps each ant takes

# Possible moves: up, down, left, right
# Each move changes x or y by +1 or -1
step_choices = np.array([(0, 1), (0, -1), (1, 0), (-1, 0)])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

final_distances = []  # Store how far each ant ends up from the start

# Colour palette so each ant gets its own distinct colour
ant_colors = plt.cm.tab10(np.linspace(0, 1, num_ants))

for ant_number in range(num_ants):
    # Start every ant at the origin (0, 0)
    x_position = 0
    y_position = 0
    path_x = [0]
    path_y = [0]

    for step in range(num_steps):
        # Pick a random direction from our 4 choices
        move = step_choices[np.random.randint(0, 4)]
        x_position += move[0]
        y_position += move[1]
        path_x.append(x_position)
        path_y.append(y_position)

    # Calculate the straight-line distance from the starting point (Pythagoras theorem)
    final_distance = math.sqrt(x_position**2 + y_position**2)
    final_distances.append(final_distance)

    # Plot this ant's entire path on the left chart
    ax1.plot(path_x, path_y, color=ant_colors[ant_number],
             alpha=0.6, linewidth=0.8, label=f'Ant {ant_number + 1}')

    # Mark the start (black dot) and end (star) positions
    ax1.scatter([0], [0], color='black', s=80, zorder=5)
    ax1.scatter([x_position], [y_position], color=ant_colors[ant_number],
                s=80, marker='*', zorder=5)

# Add dummy legend entries for start and end markers
ax1.scatter([], [], color='black', s=80, label='Start (origin)')
ax1.scatter([], [], color='gray', s=80, marker='*', label='Final position')
ax1.set_xlabel('X Position', fontsize=12)
ax1.set_ylabel('Y Position', fontsize=12)
ax1.set_title(f'Random Walk — {num_ants} Lost Ants, {num_steps:,} Steps Each',
              fontsize=13, fontweight='bold')
ax1.legend(fontsize=9, loc='upper left')
ax1.set_aspect('equal')

# --- Histogram: how far do 500 ants end up? ---
# We run a larger batch for the histogram to get a smooth distribution
num_ants_hist = 500
all_distances = []

for _ in range(num_ants_hist):
    # Generate all steps at once using numpy (much faster than a per-step loop)
    step_indices = np.random.randint(0, 4, num_steps)
    all_steps = step_choices[step_indices]
    final_pos = np.sum(all_steps, axis=0)  # Sum all steps to get final position
    distance = math.sqrt(final_pos[0]**2 + final_pos[1]**2)
    all_distances.append(distance)

ax2.hist(all_distances, bins=30, color='#2A9D8F', edgecolor='white', alpha=0.85)
ax2.axvline(x=np.mean(all_distances), color='#D64045', linestyle='--',
            linewidth=2.5,
            label=f'Average distance = {np.mean(all_distances):.1f} steps')
ax2.axvline(x=math.sqrt(num_steps), color='#1E3A5F', linestyle='-.',
            linewidth=2.5,
            label=f'Expected distance \u2248 \u221a{num_steps} = {math.sqrt(num_steps):.1f}')
ax2.set_xlabel('Final Distance from Start (steps)', fontsize=12)
ax2.set_ylabel('Number of Ants', fontsize=12)
ax2.set_title(f'How Far Do Lost Ants End Up?\n'
              f'{num_ants_hist} ants, {num_steps:,} steps each',
              fontsize=13, fontweight='bold')
ax2.legend(fontsize=11)

plt.tight_layout()
plt.show()

print(f"\nRandom Walk Results ({num_ants_hist} ants, {num_steps:,} steps each):")
print(f"  Average final distance:   {np.mean(all_distances):.2f} steps")
print(f"  Expected distance (\u221an):   {math.sqrt(num_steps):.2f} steps")
print(f"  Closest ant to origin:    {min(all_distances):.2f} steps")
print(f"  Furthest ant from origin: {max(all_distances):.2f} steps")
```

### **What do we see here?**

- Each ant's path looks **completely different** — total chaos! No two paths are alike.
- Yet on average, ants end up about **√n steps** away from the start (√1000 ≈ 31.6). Our simulation confirms this beautifully — the red dashed line and the blue dash-dot line overlap closely.
- This is called the **Square Root Rule of random walks** — beautiful mathematical order hidden inside pure chaos. The ant has no plan, no memory, no intention — yet the average outcome follows a precise formula.
- **Real-world applications of random walks:**
  - **Heat and diffusion:** how heat spreads through a metal follows the same equations as a random walk
  - **Epidemiology:** how an infectious disease spreads through a city is modelled with random walks
  - **Finance:** stock price movements are modelled as random walks (the "efficient market hypothesis")
  - **AI and robotics:** reinforcement learning agents explore their environment using random walk strategies

## Why Monte Carlo Matters — It Powers Real AI Systems

Everything you just learned is not just for maths class. These exact techniques — random sampling, running thousands of simulations, counting outcomes — are inside the AI tools you use every day. Let's look at where.

---

### 1. Reinforcement Learning and AlphaGo (Monte Carlo Tree Search)

In 2016, an AI system called **AlphaGo** (built by Google DeepMind) beat the world's best human player at the ancient board game Go. This was considered one of the biggest milestones in AI history — Go has more possible board positions than there are atoms in the observable universe, so you cannot pre-compute every outcome.

AlphaGo's secret weapon? **Monte Carlo Tree Search (MCTS)** — an algorithm that simulates millions of random game sequences from any position, counts which random "play-outs" lead to wins, and uses those statistics to decide the best move.

Think of it like this: instead of memorising every chess move ever played, you imagine randomly completing the game thousands of times from your current position and remember which starting moves tended to produce wins. That is exactly MCTS.

The same technique powers AI opponents in video games, Go, chess engines, and even AI that plays complex strategy games like StarCraft.

---

### 2. Training Neural Networks (Stochastic Gradient Descent)

**Neural networks** — the backbone of ChatGPT, image recognition, voice assistants, and almost every modern AI system — learn from data. But the datasets are enormous: billions of images, trillions of words.

Instead of training on all the data at once (which would take forever), neural networks use **Stochastic Gradient Descent (SGD)** — where "stochastic" literally means *random*. In each training step, the network picks a **random small batch** of training examples, learns from just those, then picks another random batch, and so on.

This random sampling is a form of Monte Carlo! The random batches give a "good enough" estimate of the overall pattern — just like our random dart throws gave a good enough estimate of π. Remarkably, the randomness also *helps* the learning process by preventing the AI from getting stuck in bad solutions.

Analogy: instead of tasting every single spoonful of a giant pot of soup to check the flavour, you taste a random spoonful — it gives you a good enough idea of whether it needs more salt!

---

### 3. Generative AI — Images, Music, and Text

When you ask **DALL-E** or **Stable Diffusion** to generate an image of "a cat astronaut on the moon," what actually happens?

The model starts with a picture of **pure random noise** — literally random pixels — and gradually transforms it into a coherent image through a process called **diffusion**. At each step, the model makes probabilistic (random) decisions about how to reduce the noise. Monte Carlo sampling guides which random paths to explore at each stage.

Similarly, **ChatGPT** and other language models do not "look up" answers from a database. At every word, the model calculates the probability of thousands of possible next words — and then **randomly samples** from those probabilities. This is why asking the same question twice can give slightly different answers!

---

### 4. Medical AI and Drug Discovery

Designing a new medicine is one of the hardest problems in science. A single drug molecule might have thousands of possible shapes and arrangements of atoms — and only a tiny fraction of those arrangements will actually work as a medicine.

Scientists use **Monte Carlo simulations** to randomly sample millions of possible molecular configurations and estimate which ones are likely to bind to a disease target. This has accelerated drug discovery dramatically — treatments that might have taken 20 years to find can now be identified in months.

Monte Carlo is also used in **radiation therapy planning** for cancer treatment. Doctors simulate billions of random photon paths through virtual patient anatomy to figure out exactly how to aim radiation beams to destroy tumour cells while sparing healthy tissue.

---

### 5. Autonomous Vehicles (Self-Driving Cars)

A self-driving car must make split-second decisions: should it brake? Steer left? Let the pedestrian cross?

Before taking any action, the car's AI runs **thousands of simulated scenarios** in milliseconds, each representing a slightly different possible future (the pedestrian might speed up, might stop, the car in the next lane might change lanes, etc.). Each scenario is essentially a random trial. By simulating enough random futures, the AI can estimate which action is safest across the full range of possibilities.

A car's AI might run 100,000 simulated scenarios in seconds to decide whether to brake or steer. This is Monte Carlo running in real time, in a car, at highway speed.

### Key Takeaways

1. **Monte Carlo uses random numbers and many trials** to answer hard questions without needing complex equations — you just need to count.
2. **More trials = more accurate answer** (Law of Large Numbers). This is not a hope — it is a mathematical guarantee.
3. **Sometimes random simulation reveals surprising truths** — like the Birthday Paradox showing that only 23 people are needed for a 50% chance of a shared birthday.
4. **Monte Carlo can estimate areas under curves** without any calculus — making "integration" accessible to everyone.
5. **Random walks reveal hidden mathematical patterns** — the \u221an rule shows that beautiful order can emerge from complete chaos.
6. **These exact techniques power modern AI:** game-playing AI (AlphaGo), language models (ChatGPT), image generators (DALL-E), self-driving cars, and medical AI.
7. **Setting a random seed (like 42) makes experiments reproducible** — a crucial practice in real science so that results can be verified and shared.

---

### Limitations of Monte Carlo

Monte Carlo is powerful, but not perfect:

- **It can be slow.** To get high accuracy, you often need millions of trials — which takes computing time and memory.
- **Results have some randomness.** Running the same code with a different random seed will give a slightly different answer — unlike an exact mathematical formula.
- **Not always the best method.** When a direct formula exists (like the area of a circle = \u03c0r\u00b2), using that formula is always faster and more accurate than simulating it.

Monte Carlo shines brightest when problems are **too complex for any formula** — and in the modern world, most interesting problems are exactly that.
